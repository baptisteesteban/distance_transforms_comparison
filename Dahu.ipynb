```python
def dahu_distance_transform(m: np.ndarray, M: np.ndarray) -> np.ndarray:
    deja_vu = np.zeros(m.shape, dtype=bool)
    min_img = np.empty_like(m)
    max_img = np.empty_like(m)
    F = np.empty_like(m)
    q = PQueue()

    F[0, 0] = m[0, 0]
    min_img[0, 0] = m[0, 0]
    max_img[0, 0] = m[0, 0]
    deja_vu[0, 0] = True
    q.push((0, 0), 0)

    while not q.empty():
        (l, c) = q.pop()
        d = q.distance
        for dl, dc in C4:
            nl, nc = l + dl, c + dc
            if in_domain(m.shape, nl, nc) and not deja_vu[nl, nc]:
                f = clamp(F[l, c], m[nl, nc], M[nl, nc])
                F[nl, nc] = f
                diff = abs(int(F[l, c]) - int(F[nl, nc]))
                min_img[nl, nc] = min(f, min_img[l, c])
                max_img[nl, nc] = max(f, max_img[l, c])
                deja_vu[nl, nc] = True
                q.push((nl, nc), diff)

    return max_img - min_img
```

```python
dahu = dahu_distance_transform(m, M)
```
